Importando librerías

```python
import pandas as pd
import json
```

Abriendo el archivo json de tweets

```python
with open('farmers-protest-tweets-2021-03-5.json', 'r', encoding='utf-8') as file:
    twitter = [json.loads(diccionario) for diccionario in file]
```

```python
columnas = [[d['id'], d['content'], d['renderedContent'], d['retweetCount'], d['user'], d['date']] for d in twitter]
df_twitter = pd.DataFrame(columnas)
df_twitter.columns = ['id', 'content', 'renderedContent', 'retweetCount', 'user', 'date']
```

Cortando el formato de fechas

```python
def fecha(date):
    return date[:10]
```

```python
df_twitter['date'] = df_twitter['date'].apply(fecha)
```

Top 10 más retweetiados

```python
def top_retweeted(df):
    df_retweets = df.nlargest(10, 'retweetCount')
    top = [df_retweets.content]
    for topi in top:
        print(topi)
    return 
```

Top 10 usuarios que más tweetean

```python
from collections import Counter
def top_usuarios(df):
    users = [u['username'] for u in df['user']]
    users_count = Counter(users)
    tops_u = users_count.most_common(10)
    for username in tops_u:
        print(username[0])
    return 
```

Top días más tweetiados

```python
def top_dias(df):
    #sacado de https://stackoverflow.com/questions/48590268/pandas-get-the-most-frequent-values-of-a-column
    tops_days = df['date'].value_counts()[:10].index.tolist()
    for dia in tops_days:
        print(dia)
    return 
```

Top hashtags más usados

```python
def top_hashtags(df):
    hashtags = []
    for twitter in df['content']:
        for word in twitter.split():
            if word[0] == '#':
                hashtags.append(word)
    cuenta = Counter(hashtags)
    tops = cuenta.most_common(10)
    for hashtag in tops:
        print(hashtag[0])
    return 
```

```python
def main(df):
    print("Top 10 tweets más retweeteados\n")
    top_retweeted(df)
    print("\n")
    print("Top 10 usuarios con más tweets\n")
    top_usuarios(df)
    print("\n")
    print("Top 10 días con más tweets\n")
    top_dias(df)
    print("\n")
    print("Top 10 hashtags usados\n")
    top_hashtags(df)
    print("\n")
```

```python
main(df_twitter)
```
